```python
import pandas as pd

# 1. تحميل ملف الطقس فقط
weather = pd.read_csv('/content/weather_dataset_final.csv')

# 2. تحويل عمود التاريخ لنوع Datetime لتمكين العمليات الحسابية
weather['date'] = pd.to_datetime(weather['date'])

# 3. حل مشكلة التاريخ (Time-Shift Alignment)
# سنقوم بإزاحة التواريخ من يناير 2023 [1] لتطابق بداية رحلاتك في أبريل 2026
# الإزاحة التقريبية هي 1213 يوم (أو يمكنك وضع التاريخ المستهدف بدقة)
target_start_date = pd.to_datetime('2026-04-28')
current_start_date = weather['date'].min()
time_offset = target_start_date - current_start_date

# تطبيق الإزاحة الزمنية
weather['date'] = weather['date'] + time_offset

# 4. إنشاء الـ weather_key (مهم جداً للربط في Power Query لاحقاً)
# هذا العمود سيجمع كود المطار والتاريخ الجديد ليكون مفتاحاً فريداً
weather['weather_key'] = (
    weather['iata_code'].astype(str) +
    "_" +
    weather['date'].dt.date.astype(str)
)

# 5. حفظ الملف الجديد بصيغة CSV
weather.to_csv('Weather_Clean_2026_Ready.csv', index=False)

print("Weather file fixed! ✅")
print(f"Dates shifted to start from: {weather['date'].min().date()}")
print("New file 'Weather_Clean_2026_Ready.csv' is ready for Power Query.")
```
